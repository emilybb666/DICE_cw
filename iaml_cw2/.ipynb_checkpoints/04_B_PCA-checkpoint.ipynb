## Description of the dataset

In this part of the assignment we will explore the MNIST digits dataset. We expect the digits to lie in a lower-dimensional manifold and want to examine the representation we get by applying Principal Components Analysis (PCA). PCA maps the data into a new space by effectively rotating the base vectors of the input space to the directions with the highest variance.

### Description of the Data

This MNIST Dataset is a collection of handwritten digits. The samples are partitioned (nearly) evenly across the 10 different digit classes {0, 1, . . . , 9}. We use a preprocessed version for which the data are $8 \times 8$ pixel images containing one digit each. For further details on how the digits are preprocessed, see the sklearn documentation. The images are grayscale, with each pixel taking values in {0, 1, . . . , 16}, where 0 corresponds to black (weakest intensity) and 16 corresponds to white (strongest intensity). Therefore, the dataset is a N Ã— 64
dimensional matrix where each dimension corresponds to a pixel from the image and N is the number of
images.

````python
#(a) # Your Code goes here:
pca = PCA(n_components=64, random_state=1000, svd_solver='full')
pca.fit(datas)
#X_tran = pca.transform(datas)
var_ratio = pca.explained_variance_ratio_
var_sort = sorted(var_ratio, reverse=True)

idx = sorted(range(len(var_ratio)), key=lambda k: var_ratio[k])
df_var = pd.DataFrame({'idx': idx, 'variance_ratio': var_sort}, columns=['idx', 'variance_ratio'])

fig, ax = plt.subplots(figsize=(10,5))
x_tick = arange(0,64,1)
ax.bar(x_tick, var_ratio)

ax.set_xlabel(' components', fontsize=12)
ax.set_ylabel(' cumulative explained variance', fontsize=12)

ax2 = ax.twinx()
cum = []
length = len(var_ratio)
temp = 0
for i in range(length):        
    temp += var_ratio[i] 
    cum.append(temp)


ax2.plot(x_tick, cum, '-ro')
for i in range(64):
    cum[i]=cum[i]*100
    if(round(cum[i])==80):
        point_80 = i

        


ax2.hlines(0.8, 0, 64, color='green')

ax.vlines(point_80, 0, 0.14, color='green')

plt.show()
````

````python
#(a) # Your Code goes here:
# The heatmap of the original digit
digit0 = digits.data[0,:]
plt.subplots(figsize=(14,10))
plt.subplot(2,2,1)
sns.heatmap(digit0.reshape(8,8), cmap='binary')
plt.title("The heatmap of the origin digit")

pca = PCA(n_components=64, random_state=1000, svd_solver='full')
pca.fit(datas)
X_tran = pca.transform(datas)
X_mean = pca.mean_

dot_product = np.dot(X_tran[0], comp)
recon = X_mean + dot_product
MAD = abs(digit0 - recon).mean()

plt.subplot(2,2,2)
sns.heatmap(recon.reshape(8,8), cmap='binary')
plt.title("The heatmap of the reconstracted digit")
plt.show()

print("The MAD is ", MAD)
````

````python
#(b) # Your Code goes here:
# component 0
plt.subplots(figsize=(12,15))
X_tran = pca.transform(datas)
dig0_tran = X_tran[:1]
comp_list =[1,2,5,15,30] 
length = len(comp_list)
plt.subplot(3,2,1)
sns.heatmap(dig0_tran.reshape(8,8), cmap='binary')
plt.title("The reconstruction with 0 components")

for i in range(length):
    plt.subplot(3,2,i+2)
    
    dot_product = np.dot(X_tran[:1,:comp_list[i]], comp[:comp_list[i]])
    recon = X_mean + dot_product
    MAD = abs(digit0 - recon).mean()
    sns.heatmap(recon.reshape(8,8), cmap='binary')
    plt.title("The reconstruction with {} components \n MAD is {}".format(comp_list[i], MAD),fontsize=12)


plt.show()
````

(c) ***Your answer goes here:***

In (a), we reconstracted the digit using all 64 components which is almost the same as the original one. The MAD is almost zero. In (b) we can see as the total number of components we used increaases, the digit becomes more clear. The digit is already recogniseable when 5 components are used.
